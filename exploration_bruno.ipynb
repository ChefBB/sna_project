```python
import pandas as pd

df = pd.read_csv('data.csv')
df
```

```python
duplicates = df[df.duplicated(subset=df.columns.difference(['Team']), keep=False)]
duplicates_sorted = duplicates.sort_values(by=df.columns.difference(['Team']).tolist())
duplicates_sorted
```

```python
import hypernetx as hnx

# Create a dictionary where keys are (Team, Season) pairs and values are lists of players
hyperedges = {}
for _, row in df.iterrows():
    key = (row['Team'], row['Season'])
    if key not in hyperedges:
        hyperedges[key] = []
    hyperedges[key].append(row['Player'])

# Create the hypergraph
H = hnx.Hypergraph(hyperedges)

# Display the hypergraph summary
print(H)
```
